#Generating Data

```python
import matplotlib.pyplot as plt
import numpy as np

n_samples = 25
# Generate 50 random points between 0 and 100
x = np.random.rand(n_samples) * 10

# Add random noise between -5 and 5
noise = np.random.rand(n_samples) * 15 - 5

# Calculate y values with x^2 + sin(x) and add noise
y = x**2 * np.sin(x) + noise
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Define the neural network model with increased complexity
model = tf.keras.Sequential([
  tf.keras.layers.Dense(128, kernel_initializer='normal', activation='relu', input_shape=(1,)),  # Hidden layer with 50 neurons and ReLU activation
  tf.keras.layers.Dense(256, kernel_initializer='normal', activation='relu', input_shape=(1,)),  # Hidden layer with 50 neurons and ReLU activation
  tf.keras.layers.Dense(256, kernel_initializer='normal', activation='relu', input_shape=(1,)),  # Hidden layer with 50 neurons and ReLU activation
  tf.keras.layers.Dense(256, kernel_initializer='normal', activation='relu', input_shape=(1,)),  # Hidden layer with 50 neurons and ReLU activation
  tf.keras.layers.Dense(1, kernel_initializer='normal',activation='linear')  # Output layer with 1 neuron
])

# Add L1 regularization (optional but can help prevent overfitting)
model.compile(loss='mse', optimizer=tf.keras.optimizers.AdamW(learning_rate=1e-3))

# Train the model with more epochs
model.fit(x[:, np.newaxis], y, epochs=5000, batch_size=512, verbose=1)

# Predict using the trained model
predicted_y = model.predict(x[:, np.newaxis])

# Plot the results
plt.scatter(x, y, label='Original Data')
plt.plot(x, predicted_y.squeeze(), label='Predicted Data')
plt.xlabel('x (normalized)')  # Update label for normalized data
plt.ylabel('y')
plt.legend()
plt.show()
```
